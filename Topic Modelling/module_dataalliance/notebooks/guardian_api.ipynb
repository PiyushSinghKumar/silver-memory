```python
import requests
import pprint
import pymongo
import json
```

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
Articles = myclient.guardian_articles
```

```python
collection = Articles.guardian_collection
```

```python
query = "(covid OR Covid)"
query_fields = "body"
section = "news"  # https://open-platform.theguardian.com/documentation/section
#tag = "world/extreme-weather"  # https://open-platform.theguardian.com/documentation/tag
from_date = "2022-02-01"
```

```python
page = 1
status_code = 200
```

```python
while status_code == 200:
    query_url = f"https://content.guardianapis.com/search?" \
            f"api-key={'9d3d5b03-3d13-45a1-b8c2-95de7d7da4e0'}" \
            f"&q={query}" \
            f"&query-fields={query_fields}" \
            f"&section={section}" \
            f"&from-date={from_date}" \
            f"&show-fields=body,shortUrl" \
            f"&page-size={'50'}" \
            f"&page={page}"
    #print(page)
    r = requests.get(query_url)
    if r.status_code == 200:
        out = r.json()
        collection.insert_one(out)
    #print(out)
    #respond.append(out)
    #respond['response'].append(out['response'])
    page = page + 1
    #print(page)
    #print("Ran")
    #print(r.status_code)
    status_code = r.status_code
```
